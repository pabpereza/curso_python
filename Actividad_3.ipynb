## Ejercicio 3

Define una función dentro de la clase Numero que a partir de una cadena de caracteres, devuelve True si esa cadena de caracteres corresponde con un número romano y falso en caso contrario. Después, modifica el método para que lance un error en el caso de que el valor que nos pasan por parámetro no se corresponde con el patrón de un número romano

```python

class Numero():
    
    romanos = {
        1: 'I',
        4: 'IV',
        5: 'V',
        9: 'IX',
        10: 'X',
        40: 'XL',
        50: 'L',
        90: 'XC',
        100: 'C',
        400: 'CD',
        500: 'D',
        900: 'CM',
        1000: 'M'
    }


    def __init__(self, numero):
        self.normal = numero
        self.romano = self.pasar_romano(numero, self.romanos)

    def pasar_romano(self, numero, romanos):
        resultado_romano = ""
        for valor, simbolo in sorted(romanos.items(), key=lambda x: x[0], reverse=True):
            while numero >= valor:
                resultado_romano += simbolo
                numero -= valor
        
        return resultado_romano
    
    def imprime(self):
        print(self.normal, self.romano)

    def suma_romano(self, numero_romano):

        nuevo_numero = self.convertir_decimal(numero_romano) + self.normal

        self.normal = nuevo_numero
        self.romano = self.pasar_romano(nuevo_numero, self.romanos)


    def convertir_decimal(self, numero_romano):
        decimal = 0
        romanos_a_decimal = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}

        for i in range(len(numero_romano)):
            if i > 0 and romanos_a_decimal[numero_romano[i]] > romanos_a_decimal[numero_romano[i - 1]]:
                decimal += romanos_a_decimal[numero_romano[i]] - 2 * romanos_a_decimal[numero_romano[i - 1]]
            else:
                decimal += romanos_a_decimal[numero_romano[i]]

        return decimal
    

    def is_romano(self, cadena):

        romanos_validos = {'I', 'V', 'X', 'L', 'C', 'D', 'M'}
        
        for letra in cadena:
            if letra not in romanos_validos:
                return False
        return True
```

## Ejercicio 5

En la clase MejorNumero, crea un nuevo método que reciba una lista con 3 números romanos. A continuación, iterando sobre los elementos de la lista llamará a la función suma_romano(). Los posibles errores se tendrán que gestionar con excepciones para mostrar un mensaje y seguir ejecutando el siguiente número.

```python
class MejorNumero(Numero):

    def __init__(self, numero):
        # Llama al constructor de la clase base (Numero)
        super().__init__(numero)

    def resta(self, otro_romano):
        if not self.is_romano(otro_romano):
            raise ValueError("El valor proporcionado no es un número romano válido.")

        otro_decimal = self.convertir_decimal(otro_romano)
        resultado_decimal = self.normal - otro_decimal

        # Maneja casos en los que el resultado es menor o igual a cero
        if resultado_decimal <= 0:
            raise ValueError("La resta produce un resultado no positivo en el sistema de números romanos.")

        self.normal = resultado_decimal
        self.romano = self.pasar_romano(resultado_decimal, self.romanos)


    def multiplica(self, otro_romano):
        if not self.is_romano(otro_romano):
            raise ValueError("El valor proporcionado no es un número romano válido.")


        otro_decimal = self.convertir_decimal(otro_romano)
        resultado_decimal = self.normal * otro_decimal

        # Maneja casos en los que el resultado sea demasiado grande para representar en el sistema romano
        if resultado_decimal > 3999:
            raise ValueError("El resultado de la multiplicación es mayor que 3999, el límite en el sistema de números romanos.")

        self.normal = resultado_decimal
        self.romano = self.pasar_romano(resultado_decimal, self.romanos)


    def iterar(self, lista_romanos):

        for iromano in lista_romanos:
            if not str(iromano).isnumeric():
                self.suma_romano(iromano)
            else:
                print("Ha fallado el número " + str(iromano))
    
```
